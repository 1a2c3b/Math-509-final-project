```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import ClusterCentroids
from sklearn.metrics import confusion_matrix

# --- PREPARE DATA (example placeholders) ---
# Replace X_scaled, y with your actual preprocessed data
# X_scaled = ...
# y = ...

# We'll do one single train/test split at the start
X_train_base, X_test_base, y_train_base, y_test_base = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# --- DEFINE SAMPLING TECHNIQUES ---
def get_resampled_data(sampling: str):
    """
    Return (X_res, y_res) for the chosen sampling method.
    'none'  : no sampling
    'smote' : SMOTE oversampling
    'under' : undersampling (e.g., K-Means or ClusterCentroids)
    """
    if sampling == 'none':
        return X_train_base, y_train_base
    elif sampling == 'smote':
        sm = SMOTE(random_state=42)
        return sm.fit_resample(X_train_base, y_train_base)
    elif sampling == 'under':
        # Example: use ClusterCentroids (K-Means based undersampling)
        cc = ClusterCentroids(random_state=42)
        return cc.fit_resample(X_train_base, y_train_base)
    else:
        raise ValueError(f"Unknown sampling method: {sampling}")

# --- DEFINE MODELS ---
def get_model(model_name: str):
    """
    Return an untrained model corresponding to the model_name.
    'logreg'  : Logistic Regression
    'rf'      : Random Forest
    'xgb'     : XGBoost
    'knn'     : KNN
    'ensemble': VotingClassifier of the above
    """
    if model_name == 'logreg':
        return LogisticRegression(max_iter=1000, random_state=42)
    elif model_name == 'rf':
        return RandomForestClassifier(n_estimators=100, random_state=42)
    elif model_name == 'xgb':
        return XGBClassifier(
            n_estimators=100,
            max_depth=4,
            learning_rate=0.1,
            use_label_encoder=False,
            eval_metric='logloss',
            random_state=42
        )
    elif model_name == 'knn':
        return KNeighborsClassifier(n_neighbors=5)
    elif model_name == 'ensemble':
        # Create an ensemble with 4 base estimators
        est = [
            ('lr', LogisticRegression(max_iter=1000, random_state=42)),
            ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
            ('xgb', XGBClassifier(
                n_estimators=50,
                max_depth=3,
                learning_rate=0.1,
                use_label_encoder=False,
                eval_metric='logloss',
                random_state=42)
            ),
            ('knn', KNeighborsClassifier(n_neighbors=5))
        ]
        return VotingClassifier(estimators=est, voting='soft')
    else:
        raise ValueError(f"Unknown model name: {model_name}")

# --- PLOT FUNCTION ---
def train_and_plot_hist(
    ax, sampling: str, model_name: str,
    threshold_mask_low=0.01, threshold_mask_high=1.0
):
    """
    1) Resample data according to 'sampling'
    2) Build the model 'model_name'
    3) Train model
    4) Predict probabilities on X_test_base
    5) Filter out predicted probs outside [threshold_mask_low, threshold_mask_high]
    6) Plot two histograms (legit vs fraud) on the provided Axes 'ax'
    """
    X_train_res, y_train_res = get_resampled_data(sampling)
    model = get_model(model_name)

    # Train
    model.fit(X_train_res, y_train_res)

    # Predict probabilities (for the positive class)
    y_probs = model.predict_proba(X_test_base)[:, 1]

    # Filter out extreme probabilities < 0.01 or > 1.0
    mask = (y_probs >= threshold_mask_low) & (y_probs <= threshold_mask_high)
    filtered_probs = y_probs[mask]
    filtered_labels = y_test_base[mask]

    # Plot the histogram
    ax.hist(
        filtered_probs[filtered_labels == 0], bins=50,
        alpha=0.6, density=True, label="Legit"
    )
    ax.hist(
        filtered_probs[filtered_labels == 1], bins=50,
        alpha=0.6, density=True, label="Fraud"
    )
    ax.set_xlim(threshold_mask_low, threshold_mask_high)
    ax.set_title(f"{sampling.upper()} + {model_name.upper()}")
    ax.set_xlabel("Predicted Probability")
    ax.set_ylabel("Density")
    ax.grid(True)

# --- MAIN SCRIPT: CREATE 3Ã—5 SUBPLOTS ---
sampling_methods = ['none', 'smote', 'under']
models = ['logreg', 'rf', 'xgb', 'knn', 'ensemble']

fig, axes = plt.subplots(
    nrows=3, ncols=5, figsize=(22, 12),
    sharex=False, sharey=False
)

for i, sampling in enumerate(sampling_methods):
    for j, model_name in enumerate(models):
        ax = axes[i, j]
        train_and_plot_hist(
            ax, sampling, model_name,
            threshold_mask_low=0.01, threshold_mask_high=1.0
        )
        # Show legend only in the first row & first column (optional)
        if i == 0 and j == 0:
            ax.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import ClusterCentroids

# --- Define your baseline train/test split (Replace with your actual data) ---
# For example:
# X_scaled = ...  # your normalized feature matrix (Pandas DataFrame)
# y = ...         # your target variable (Pandas Series)
# Here we assume they are already loaded.

X_train_base, X_test_base, y_train_base, y_test_base = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42)

# --- Sampling Helper Function ---
def get_resampled_data(sampling: str):
    """
    Returns resampled training data based on sampling method.
    Options:
      'none': no sampling,
      'smote': apply SMOTE,
      'under': apply undersampling using ClusterCentroids.
    """
    if sampling == 'none':
        return X_train_base, y_train_base
    elif sampling == 'smote':
        sm = SMOTE(random_state=42)
        return sm.fit_resample(X_train_base, y_train_base)
    elif sampling == 'under':
        cc = ClusterCentroids(random_state=42)
        return cc.fit_resample(X_train_base, y_train_base)
    else:
        raise ValueError(f"Unknown sampling method: {sampling}")

# --- Model Helper Function ---
def get_model(model_name: str):
    """
    Returns an untrained model based on the given model_name.
    Options:
      'logreg': Logistic Regression,
      'rf': Random Forest,
      'xgb': XGBoost,
      'knn': K-Nearest Neighbors,
      'ensemble': VotingClassifier with the four base models.
    """
    if model_name == 'logreg':
        return LogisticRegression(max_iter=1000, random_state=42)
    elif model_name == 'rf':
        return RandomForestClassifier(n_estimators=100, random_state=42)
    elif model_name == 'xgb':
        return XGBClassifier(
            n_estimators=100,
            max_depth=4,
            learning_rate=0.1,
            use_label_encoder=False,
            eval_metric='logloss',
            random_state=42
        )
    elif model_name == 'knn':
        return KNeighborsClassifier(n_neighbors=5)
    elif model_name == 'ensemble':
        estimators = [
            ('lr', LogisticRegression(max_iter=1000, random_state=42)),
            ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
            ('xgb', XGBClassifier(
                n_estimators=50,
                max_depth=3,
                learning_rate=0.1,
                use_label_encoder=False,
                eval_metric='logloss',
                random_state=42
            )),
            ('knn', KNeighborsClassifier(n_neighbors=5))
        ]
        return VotingClassifier(estimators=estimators, voting='soft')
    else:
        raise ValueError(f"Unknown model name: {model_name}")

# --- Function to Train a Model and Plot Its ROC Curve ---
def train_and_plot_roc(ax, sampling: str, model_name: str):
    # Get resampled data
    X_train_res, y_train_res = get_resampled_data(sampling)
    # Initialize and train the model
    model = get_model(model_name)
    model.fit(X_train_res, y_train_res)
    # Predict probabilities for the positive class
    y_probs = model.predict_proba(X_test_base)[:, 1]

    # Compute ROC curve and AUC
    fpr, tpr, _ = roc_curve(y_test_base, y_probs)
    auc_val = roc_auc_score(y_test_base, y_probs)
    # Plot ROC curve
    ax.plot(fpr, tpr, linewidth=2, label=f"AUC = {auc_val:.2f}")
    ax.plot([0, 1], [0, 1], "k--", linewidth=1)
    ax.set_xlim([0, 1])
    ax.set_ylim([0, 1])
    ax.set_title(f"{sampling.upper()} + {model_name.upper()}")
    ax.set_xlabel("False Positive Rate")
    ax.set_ylabel("True Positive Rate")
    ax.grid(True)

    return auc_val

# --- Define Sampling Methods and Models ---
sampling_methods = ['none', 'smote', 'under']
model_names = ['logreg', 'rf', 'xgb', 'knn', 'ensemble']

# Prepare a dictionary to store AUC values
auc_table = {}

# Create a 3 x 5 grid for ROC curves
fig, axes = plt.subplots(nrows=3, ncols=5, figsize=(22, 12))
for i, sampling in enumerate(sampling_methods):
    for j, model_name in enumerate(model_names):
        ax = axes[i, j]
        auc_val = train_and_plot_roc(ax, sampling, model_name)
        auc_table[(sampling, model_name)] = auc_val

plt.tight_layout()
plt.show()

# Create a DataFrame for the AUC table and display it
auc_df = pd.DataFrame(index=sampling_methods, columns=model_names)
for (sampling, model_name), auc in auc_table.items():
    auc_df.loc[sampling, model_name] = auc

print("AUC Table:")
print(auc_df)
```
